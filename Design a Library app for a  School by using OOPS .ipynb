```python
"""
Create a library class 

it inclues 
1. Display books
2. Add book
3. Lend book
4. Return book
5. Delete book

# PENDING WORK
# i want to use getter and setter method here

6. admin registration and update profile
7. user registration and update details


"""



class Library:    
    def __init__(self,name,books):
        self.library_name=name
        self._book_lists=books
        self.lendbook={}     
        
       
    @property            # getter method
    def book_lists(self):
        print("\nList of books available Currently in library\n")
        for no, book in enumerate(self._book_lists):
            print(no+1,".",book)
            
    @book_lists.setter              # setter method 
    def book_lists(self, new_book):
        if new_book.upper() not in self._book_lists:
            self._book_lists.append(new_book.upper())
            print("\n!! Book has been added successfully !!\n")
        else:
            print("\n!! The Book is already available in our library !!\n")
            
                                          
    def delete_book(self,dele_book):
        if dele_book.upper() in self._book_lists:
            self._book_lists.remove(dele_book.upper())
            print("\n!! Book has been remove successfully !!\n")
        else:
            print("\n!! Not a valid book !!\n")
        
    
    def lend_book(self,book,name ):
        if book.upper() in self._book_lists:
            if book.upper() not in self.lendbook.keys():
                self.lendbook.update({book.upper():name})
                print("\n!! Congratulations, you got the book !!")
            else:
                print(f"\n!! You can't take the book as it already taken by {self.lendbook[book.upper()]} !!")
        else:
            print("!! Sorry, Currently we don't have this book !!\n")
            
    def returnbook(self, book):
        if book.upper() not in self.lendbook.keys():
            print("\n!! Sorry this is not our book !!\n")
        else:
            self.lendbook.pop(book.upper())
            print("\n!! you have return the book successfully !!\n")
    
    
if __name__=='__main__':
    try:
        obj=Library("NIROB",["PYTHON","C++","JAVA","RICH DAD POOR DAD"])
        print(f"WELCOME TO {obj.library_name}")

        while True:
            print("\n1. Display books")
            print("2. Add book")
            print("3. Lend book")
            print("4. Return book")
            print("5. Delete book")
            print("6. Exit")
            x= input()
            if x=="1":
                obj.book_lists
            elif x=="2":
                new_book = input("Enter the book name you want to add : \n")
                obj.book_lists=new_book

            elif x=="3":
                book=input("Enter the book name you want to lend : \n")
                name=input("Enter your name : \n")
                obj.lend_book(book,name )
            elif x=="4":
                book= input("Enter the book name you want to return : \n")
                obj.returnbook(book)
            elif x=="5":
                dele_book=input("Enter the name of the book :\n")
                z=obj.delete_book(dele_book)
                del z
            elif x=="6":
                break
            else:
                print("\n!! Not a valid option !!")
    except Exception as e:
        print(e.args)
        print("\n!! SOMETHING WENT WRONG TRY AGAIN !!")
        
print('!! Thank you !!')
```
